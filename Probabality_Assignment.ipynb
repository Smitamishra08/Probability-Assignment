# 1. Coin Toss & Dice Simulation

```python
import random

# Coin Toss Simulation
heads = 0
tails = 0
trials = 10000

for i in range(trials):
    result = random.choice(['H', 'T'])
    if result == 'H':
        heads += 1
    else:
        tails += 1

print("Experimental Probability of Heads:", heads/trials)
print("Experimental Probability of Tails:", tails/trials)

# Dice Simulation (Sum = 7)
count_sum7 = 0

for i in range(trials):
    d1 = random.randint(1, 6)
    d2 = random.randint(1, 6)

    if d1 + d2 == 7:
        count_sum7 += 1

print("Probability of getting sum 7:", count_sum7/trials)
```

### In this experiment, I simulated tossing a coin 10,000 times using python module. I counted how
### many times heads and tails occurred & calculated their probabilities.

### Similarly, I simulated rolling two dice & checked how often the sum equals 7. The probability was
### calculated by dividing successfull outcomes by totals trails.

### This demonstrates experimental probability,which approaches theoretical probability as trails increase.

# 2. Probability of At least one "6" in 10 Rolls.

```python
import random

def probability_of_six(trials=10000):
    success = 0

    for i in range(trials):
        rolls = [random.randint(1, 6) for i in range(10)]

        if 6 in rolls:
            success += 1

    return success / trials

print("Probability of at least one 6:", probability_of_six())
```

### In this task, I simulated rolling a die 10 times and checked whether at least one roll resulted in a 6.

### If a 6 appeared in any of the 10 rolls, it was considered a successful trial. The probability was estimated by dividing the number of successful trials by the total number of trials.

### This helps in understanding probability over multiple independent events.

# 3. Conditional Probability & Bayes' Theorem

```python
import random

balls = ['R']*5 + ['G']*7 + ['B']*8
trials = 1000

prev_blue = 0
blue_to_red = 0

previous = random.choice(balls)

for i in range(trials):
    current = random.choice(balls)

    if previous == 'B':
        prev_blue += 1
        if current == 'R':
            blue_to_red += 1

    previous = current

if prev_blue > 0:
    print("P(Red | Previous Blue):", blue_to_red / prev_blue)
```

### In this simulation, I created a bag containing red, green, and blue balls. A ball was drawn randomly and replaced back into the bag.

### I tracked cases where the previous ball was blue and calculated the probability that the next ball drawn was red.

### This demonstrates conditional probability, where the probability of an event depends on a previous event. The results also align with Bayes’ theorem concepts.

# 4. Discrete Random Variable

```python
import numpy as np

values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.4]

sample = np.random.choice(values, size=1000, p=probabilities)

print("Mean:", np.mean(sample))
print("Variance:", np.var(sample))
print("Standard Deviation:", np.std(sample))
```

### Here, I generated 1000 samples from a discrete random variable using given probabilities.

### Using NumPy functions, I calculated:

- Mean (average value)

- Variance (spread of data)

- Standard deviation (dispersion)

### This helps in understanding how probability distributions behave in real data.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

data = np.random.exponential(scale=5, size=2000)

# Histogram
plt.hist(data, density=True)

# PDF Curve
x = np.linspace(0, max(data), 100)
plt.plot(x, expon.pdf(x, scale=5))

plt.title("Exponential Distribution")
plt.show()
```

### In this task, I generated 2000 random samples from an exponential distribution with a mean of 5.

### I visualized the data using:

- Histogram (to show frequency distribution)

- Probability Density Function (PDF)

### This helps to understand continuous probability distributions and their shapes.

# 6. Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt

# Step1: Generate uniform data
data = np.random.uniform(0, 1, 10000)

# Step2: Sample means
sample_means = []

for i in range(1000):
    sample = np.random.choice(data, size=30)
    sample_means.append(np.mean(sample))

# Step3: Plot Distribution
plt.hist(data, density=True)
plt.title("Uniform Distribution")
plt.show()

plt.hist(sample_means, density=True)
plt.title("Sample Mean Distribution")
plt.show()
```

### In this experiment, I generated random numbers from a uniform distribution and then created multiple samples of size 30.

### For each sample, I calculated the mean and plotted the distribution of these sample means.

### The result shows that even though the original data is uniformly distributed, the sample means follow a normal distribution. This demonstrates the Central Limit Theorem.
